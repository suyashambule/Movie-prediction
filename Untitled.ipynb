```python
import ast
```

```python
def easy(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movie['genres']=movie['genres'].apply(easy)
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(easy)
```

```python
def easy2(obj):
    l=[]
    n=0;
    for i in ast.literal_eval(obj):
        if n!=3:
            l.append(i['name'])
            n+=1
        else:
            break
    return l

```

```python
movie['cast']=movie['cast'].apply(easy2)
```

```python
def direc(obj):
    l = []
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)  #
    for i in obj:
        if i['job'] == 'Director':
            l.append(i['name'])
    return l


```

```python
movie.head()
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie.head()
```

```python
movie['tags']=movie['genres']+movie['overview']+movie['crew']+movie['cast']+movie['keywords']
```

```python
movie.head()
```

```python
movies=movie[['movie_id','title','tags']]
```

```python
movies.head()
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies['tags']=movies['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(movies['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

# Assuming each element in the 'tags' column is a list of tags
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi=cosine_similarity(vector)
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distance=simi[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(movies.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```
